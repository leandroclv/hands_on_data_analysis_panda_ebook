```python
df.loc[df['mag'] >= 7, ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
```

```python
df.loc[(df['tsunami'] == 1) & (df['alert'] == 'red'), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']] #operador & == and == e
```

```python
df.loc[(df['tsunami'] == 1) | (df['alert'] == 'red'), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']] #operador | == or == ou
```

```python
df.loc[(df['place'].str.contains('Alaska')) & (df['alert'].notnull()), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']] 
#Seleciona todos os terremotos no Alasca onde temos um valor diferente de nulo.
```

```python
df.loc[df['mag'].between(6.5, 7.5), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']] #Utilizou o método between() entre um valor e outro
```

```python
df.loc[df['magType'].isin(['mw', 'mwb']), ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
#Utilizando o método isin() para selecionar por serie booleana os valores que possuem mw e mwb na coluna magType e mostras as colunas 'alert', 'mag', 'magType', 'title', 'tsunami', 'type' das linhas dos valores True para a presença de mw e mwb 
```

```python
df.loc[[df['mag'].idxmin(), df['mag'].idxmax()], ['alert', 'mag', 'magType', 'title', 'tsunami', 'type']]
#selecionar o índice do valor mínimo e valor máximo e mostrando as colunas 'alert', 'mag', 'magType', 'title', 'tsunami', 'type'
```
